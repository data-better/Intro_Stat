```
x=c(163,   161,   168,   161,   157,   162,   153,   159,   164,   170,
    152,   160,   157,   168,   150,   165,   156,   151,   162,   150,
    156,   152,   161,   165,   168,   167,   165,   168,   159,   156)
mean(x)
sd(x)
sd(x)/sqrt(length(x))
```

* 표본 크기가 커질수록 표본 평균이 정규 분포를 따름

```
samp_5 = NULL;
for (i in 1:10000)
{
  samp_5[i] = mean(rchisq(5, 5));
}
samp_10 = NULL;
for (i in 1:10000)
{
	samp_10[i] = mean(rchisq(10, 5));
}
samp_30 = NULL;
for (i in 1:10000)
{
	samp_30[i] = mean(rchisq(30, 5));
}
par(mfrow = c(2, 2))
hist(rchisq(10000, 5), main="population dist", col="gray")
hist(samp_5, main="sample size = 5", xlab = "mean(rchisq(5, 5))", col="green")
hist(samp_10, main="sample size = 10", xlab = "mean(rchisq(10, 5))", col="blue")
hist(samp_30, main="sample size = 30", xlab = "mean(rchisq(30, 5))", col="red")
```


## 11강 두집단의 비교 실습
* 독립적인 두 집단에 신약과 위약을 제공하고, 신약의 효과로 환자의 회복기간을 측정한 데이터에 대한 t-test

```
x = c(15, 10, 13, 7, 9, 8, 21, 9, 14, 8)  # 신약 투약 집단
y = c(15, 14, 12, 8, 14, 7, 16, 10, 15, 12)  # 위약 투약 집단
t.test(x,y,alt="less",var.equal=TRUE)
t.test(x,y,alt="less")
```

## 4. 12강 회귀분석

```
x <- c(85, 90, 88, 80, 97, 75, 70, 92, 95, 78)
y <- c(90, 88, 85, 82, 98, 80, 75, 90, 92, 80)
plot(x,y)
g <- lm(y~x)
summary(g)
names(g)
resid <- g$residuals/sd(g$residuals)
plot(x,resid)
abline(h=0)
```

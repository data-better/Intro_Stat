## 1. 자장면, 우동, 짬뽕, 기타 중 메뉴 선호도를 나타내는 원 그래프 그리기




```
# 데이터 정리
prod <- c(3, 4, 1, 1, 3, 4, 3, 3, 1, 3, 2, 1, 2, 1, 2, 3, 2, 3, 1, 1, 1, 1, 4, 3, 1)
prod.counts = table(prod)
names(prod.counts) = c("자장면","우동", "짬뽕","기타")
prod.counts/25*100
pie(prod.counts,col=c("purple","green2","cyan","white"))
```

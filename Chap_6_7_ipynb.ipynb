## 1. 주사위 던지기 실험




```
RollDie = function(n) {
    sample(1:6, n, replace = T)
}
par(mfrow = c(3, 2))
r10 = RollDie(10)
hist(r10, br = c(0, 1, 2, 3, 4, 5, 6), main = "n=10", freq = F, ylab = "상대도수", 
    ylim = c(0, 0.5))
r50 = RollDie(50)
hist(r50, br = c(0, 1, 2, 3, 4, 5, 6), main = "n=50", freq = F, ylab = "상대도수", 
    ylim = c(0, 0.5))
r100 = RollDie(100)
hist(r100, br = c(0, 1, 2, 3, 4, 5, 6), main = "n=100", freq = F, ylab = "상대도수", 
    ylim = c(0, 0.5))
r1000 = RollDie(1000)
hist(r1000, br = c(0, 1, 2, 3, 4, 5, 6), main = "n=1000", freq = F, ylab = "상대도수", 
    ylim = c(0, 0.5))
r10000 = RollDie(10000)
hist(r10000, br = c(0, 1, 2, 3, 4, 5, 6), main = "n=10000", freq = F, ylab = "상대도수", 
    ylim = c(0, 0.5))
r100000 = RollDie(1e+05)
hist(r100000, br = c(0, 1, 2, 3, 4, 5, 6), main = "n=100000", freq = F, ylab = "상대도수", 
    ylim = c(0, 0.5))
```

## 2.B(10, 0.5) 확률 분포

```
n=10
x=seq(0,n)
y=dbinom(x,n,1/2)
z=numeric(n+1)
plot(x,y, main=expression("Probability of" ~ X ~ "~" ~ Bin(10,1/2)), lab = c(10, 2, 0.1), pch=15, col="gray", cex=2)
for (i in 1:(n+1))
{
  segments(x[i],y[i],x[i],z[i], lwd=14, col="gray")
}
```

## 5. 세 가지 표준편차에 따른 정규분포 그래프

```
x=seq(0,30,0.01)
mu=15;sigma=2;
y=dnorm(x,mu,sigma)
plot(x,y,type="l",lwd=2)
sigma=4
y=dnorm(x,mu,sigma)
lines(x,y,lwd=2,col=2)
sigma=6
y=dnorm(x,mu,sigma)
lines(x,y,lwd=2,col=4)
```

## 6. B(25, 0.4)를 정규분포로 근사

```
n=25; p=0.4
x=rbinom(1000,n,p)
hist(x,probability=T)
x=seq(-1,n,0.01)
y=dnorm(x,n*p,sqrt(n*p*(1-p)))
lines(x,y,lwd=2,col=3)
```

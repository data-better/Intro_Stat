## 4. 분산, 표준편차 구하기

```
# 데이터
dd = c(-100, -100, 0, 105, 105)
# 분산수식과 분산
sum((dd - 2)^2)/4
var(dd)
# 표준편차 수식과 표준편차
sqrt(sum((dd -2)^2)/4)
sd(dd)
```

## 5. 제1사분위수, 제3사분위수, 사분위수 범위 구하기 

```
data=c(10, 17, 18, 25, 28, 28)
summary(data)
quantile(data,c(.25,.75))
quantile(data,.75) - quantile(data,.25)
sd(data)
```

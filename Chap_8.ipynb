## 2. 카이제곱 분포를 이용한 계산

```
qchisq(1-0.05,9)
qchisq(0.01,20)
```

* 표본 크기가 커질수록 표본 평균이 정규 분포를 따름

```
samp_5 = NULL;
for (i in 1:10000)
{
  samp_5[i] = mean(rchisq(5, 5));
}
samp_10 = NULL;
for (i in 1:10000)
{
	samp_10[i] = mean(rchisq(10, 5));
}
samp_30 = NULL;
for (i in 1:10000)
{
	samp_30[i] = mean(rchisq(30, 5));
}
par(mfrow = c(2, 2))
hist(rchisq(10000, 5), main="population dist", col="gray")
hist(samp_5, main="sample size = 5", xlab = "mean(rchisq(5, 5))", col="green")
hist(samp_10, main="sample size = 10", xlab = "mean(rchisq(10, 5))", col="blue")
hist(samp_30, main="sample size = 30", xlab = "mean(rchisq(30, 5))", col="red")
```

## 3. t분포

* 표본 크기가 커질수록 정규 분포에 근사

```
x=seq(-5,5,0.01)
y=dnorm(x,0,1)
plot(x,y,type="l",lwd=2,col=2)
y=dt(x,2)
lines(x,y,lty=3,lwd=2,col=2)
y=dt(x,9)
lines(x,y,lty=2,lwd=2,col=2)
```

* 두 종류의 t분포와 정규 분포를 이용하여 계산한 결과 비교 

```
qt(1-0.025,8)
qt(1-0.025,120)
qnorm(1-0.025,0,1)
```

* t분포(df = 3)와 정규 분포 비교

```
x=rnorm(3000,1,1)
xbar=numeric(750)
for (i in 1:750)
 {
  j=4*(i-1)
  temp=c(x[j+1],x[j+2],x[j+3],x[j+4])
  xbar[i]=(mean(temp)-1)/sqrt(var(temp)/4)
 }
hist(xbar,probability=T,ylim=c(0,0.5),nclass=30,xlim=c(-7,7))
x1=seq(min(xbar),max(xbar),0.01)
z1=dnorm(x1,0,1)
lines(x1,z1,lwd=3,col=2)
y1=dt(x1,3)
lines(x1,y1,lwd=3,col=4)
legend("topright",legend=c("Standard Normal", "t-dist with d.f=3"),lwd=c(3,3,3),col=c(2,4))
```

## 4. 중심극한정리

```
samp_100 = NULL;
n=100;
p=0.5;
for (i in 1:10000)
{
  samp_100[i] = mean(rbinom(n,1,p));
}

z=(samp_100-p)/sqrt(p*(1-p)/n)

hist(z, main="표본의 크기 = 50",nclass=30, xlab = "표본비율", col="green",probability=T,ylim=c(0,1))

x=seq(-3,3,0.01)
y=dnorm(x,0,1)
lines(x,y,lwd=3,col=2)
```
